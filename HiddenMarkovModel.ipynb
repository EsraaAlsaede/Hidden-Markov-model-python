# Intilize value

```python
#intilazie Transtion Prob
sunny_sunny=0.8
sunny_rainy=0.2
rainy_sunny=0.4
rainy_rainy=0.6
#emission prob
sunny_happy=0.8
sunny_sad=0.2
rainy_happy=0.4
rainy_sad=0.6
#intial prob
sunny=0.6
rainy=0.4
```

# sequence value

```python
moods=['G','G','H','H','H','H','G']
prob=[]
weather=[]
 
if moods[0]=='G':
    prob.append((sunny*sunny_sad,rainy*rainy_sad))
else:
    prob.append((sunny*sunny_happy,rainy*rainy_happy))
```

# function

```python
for i in range(1,len(moods)):
    yesterday_sunny,yesterday_runny= prob[-1]
    if moods[i]=='G':
        today_sunny=max(yesterday_sunny*sunny_sunny*sunny_sad,yesterday_runny*rainy_sunny*sunny_sad) 
        today_runny=max(yesterday_sunny*sunny_rainy*rainy_sad,yesterday_runny*rainy_rainy*rainy_sad)
        prob.append((today_sunny,today_runny))
    else:
        today_sunny=max(yesterday_sunny*sunny_sunny*sunny_happy,yesterday_runny*rainy_sunny*sunny_happy)
        today_runny=max(yesterday_sunny*sunny_rainy*rainy_happy,yesterday_runny*rainy_rainy*rainy_happy)
        prob.append((today_sunny,today_runny))
for p in prob :
    if p[0]>p[1]:
        weather.append('Sanny')
    else:
        weather.append('Rainy')
print ("sequence Weather:",weather)
```
